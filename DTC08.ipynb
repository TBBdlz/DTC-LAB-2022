```python
# 4.1 Huffman coding
from sage.coding.source_coding.huffman import Huffman, frequency_table
str1="There once was a french fry"
print("The string: ",str1)
print()
ft = frequency_table(str1)
print("Frequency and probability of each letter in string: ")
Sft=sorted(ft.items())
for j in range(len(ft)):
	print(Sft[j],", P(",Sft[j][0],")=",QQ(Sft[j][1])/len(str1))
h1 = Huffman(str1)
print()
print("Codeword of each letter is:")
for letter, code in sorted(h1.encoding_table().items()):
	print("'{}' : {}".format(letter, code))
```

```python
# bit string ที่ส่งออก
bit_string = "".join(h1.encode(letter) for letter in str1)
print(bit_string)
```

```python
# entropy = Sigma p_i log(1 / p_i)
from math import log
total_count = sum(ft.values())
entropy = 0

for count in ft.values():
    p = count / total_count
    entropy -= p * log(p)

print(f'H(X) = {entropy:.6f}')
```

```python
total_count = sum(ft.values())
average_length = 0

for count in ft.values():
	p = count / total_count
	average_length += p * count

print(f'Average length is {average_length:.6f}')
```

```python
# efficiency coefficient eta
eta = entropy / average_length
print(f'Eta = {eta:.6f}')
```
