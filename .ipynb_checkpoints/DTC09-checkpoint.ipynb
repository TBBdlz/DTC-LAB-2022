# 3 การทดลอง

## 3.1 ขาออกของชุดตัวสหสัมพันธ์เมื่อช่องสัญญาณไม่เจือปนสัญญาณรบกวน
### 1. เขียนโปรแกรมด้วยจูเลียเพื่อนิยามฟังก์ชันดังรูปที่ 3 ด้วยคําสั่งต่อไปนี้ ตามด้วย shift+enter

```julia
# define signal s0(t) and s1(t)
function s0(t)
    if (t>=0) & (t<=Tb)
        return 1
    else
        return 0
    end
end;
function s1(t)
    if (t>=0) & (t<=Tb/2)
        return 1
    elseif (t>Tb/2) & (t<=Tb)
        return -1
    else
        return 0
    end
end;
```

### 2.เขียนโปรแกรมด้วยจูเลียเพื่อพล็อตกราฟรูปคลื่นสัญญาณ s0(t) ด้วยคําสั่งต่อไปนี้ ตามด้วย shift+enter

```julia
# Plot signal s0(t)
using Plots
using LaTeXStrings
plotly()
Tb = 1
plot(s0, -0.5, 1.5,xlim=(-0.5,1.5),legend=:none)
plot!(ylabel="s₀t)",xlabel="t")
```

### 3. เขียนโปรแกรมด้วยจูเลียเพื่อพล็อตกราฟรูปคลื่นสัญญาณ s1(t) ด้วยคําสั่งต่อไปนี้ ตามด้วย shift+enter

```julia
# Plot signal s1(t)
using Plots
using LaTeXStrings
plotly()
Tb=1
plot(s1, -0.5, 1.5,xlim=(-0.5,1.5),legend=:none)
plot!(ylabel="s₁(t)",xlabel="t")
```

### 4. บันทึกและสังเกตุผล

### 5. เขียนโปรแกรมด้วยจูเลียเพื่อแสดงเวกเตอร์สองค่าที่ขาออกของชุดตัวสหสัมพันธ์เมื่อส่งรูปคลื่นสัญญาณ s0(t) และ s1(t) สมมติช่องสัญญาณไม่เจือปนสัญญาณรบกวน ด้วยคําสั่งต่อไปนี้ ตามด้วย shift+enter

```julia
# define r(t)*s0(t)
# and r(t)*s1(t)
function mult_rs0(t)
    return r(t)*s0(t)
end;

function mult_rs1(t)
    return r(t)*s1(t)
end;

# vector represenation of s0(t)*s0(t)
# quadgk(args) is numerical integral function
using QuadGK
Tb=1
r(t)=s0(t) # transmit s0(t)
r0=quadgk(mult_rs0, 0, Tb)[1]
r1=quadgk(mult_rs1, 0, Tb)[1]
plot((r0,r1),markershape=:circle,markersize=8,legend=:none,xlim=(-0.1,1.1))
r(t)=s1(t) # transmit s1(t)
r0=quadgk(mult_rs0, 0, Tb)[1]
r1=quadgk(mult_rs1, 0, Tb)[1]
plot!((r0,r1),markershape=:circle,markersize=8,legend=:none,xlim=(-0.1,1.1));
plot!(xlabel="r₀",ylabel="r₁")
x_bound = range(0,stop=1,length=100)
y_bound = x_bound
plot!(y_bound,x_bound,color=:green)
```

```julia
# 3.2 ขาออกของชุดตัวสหสัมพันธ์เมื่อช่องสัญญาณมีเฉพาะสัญญาณรบกวน
# 8. เขียนโปรแกรมด้วยจูเลียเพื่อแสดงเวกเตอร์ของสัญญาณรบกวนด้วยความแปรปรวน σ2 = 0.25 (เท่ากับค่ากําลัง) ที่ขาออกของชุดตัวสหสัมพันธ์ สมมติไม่มีสารส่งออกช่องสัญญาณ ด้วยคําสั่งต่อไปนี้ ตามด้วย shift+enter
# Simulate 100 values of noise vectors (n0,n1)
# using Gaussian RV,
# where mean=0, variance=N0/2=0.25
#
using Plots
using LaTeXStrings
plotly()
mean=0
sigma=0.5
n0=sigma*randn(100)
n1=sigma*randn(100)
plot(n0,n1,seriestype=:scatter,title="100 values of noise vector, n")
plot!(xlabel="n₀",ylabel="n₁",legend=:none)
plot!(xlim=(-1.5,1.5),ylim=(-1.5,1.5))
x_bound = range(-1.5,stop=1.5,length=100)
y_bound = x_bound
plot!(y_bound,x_bound,color=:green)
```

```julia
# 3.3 ขาออกของชุดตัวสหสัมพันธ์เมื่อส่งรูปคลื่นของสัญญาณผ่านช่องสัญญาณเพิ่มการรบกวน
# 14. เขียนโปรแกรมด้วยจูเลียเพื่อแสดงเวกเตอร์สองค่าที่ขาออกของชุดตัวสหสัมพันธ์เมื่อส่งรูปคลื่นสัญญาณ s0(t) และ s1(t) สมมติช่อง
# สัญญาณเจือปนสัญญาณรบกวนด้วยความแปรปรวน σ2 = 0.01 (เท่ากับค่ากําลัง) ด้วยคําสั่งต่อไปนี้ ตามด้วย shift+enter

# Simulate 100 values of recieved vectors (r0,r1)
# that equal to s_i+n, where n is Gaussian RV,
# mean=0, variance=N0/2=0.01
#
using Plots
using LaTeXStrings
plotly()
mean=0
sigma=0.1
vs0=[1 0]
vs1=[0 1]
n0=sigma*randn(100)
n1=sigma*randn(100)
n2=sigma*randn(100)
n3=sigma*randn(100)
x1=vs0[1].+n0
y1=n1
x2=n2
y2=vs1[2].+n3
plot(x1,y1,seriestype=:scatter,title="100 values of received vector, r",markershape=:circle,markersize=4)
plot!(x2,y2,seriestype=:scatter,markershape=:rect,markersize=4)
plot!(x2,y2,seriestype=:scatter,markershape=:rect,markersize=4)
plot!(xlabel="r₀",ylabel="r₁",legend=:none)
plot!(xlim=(-1.5,1.5),ylim=(-1.5,1.5))
x_bound = range(-1.5,stop=1.5,length=100)
y_bound = x_bound
plot!(y_bound,x_bound,color=:green)
```
