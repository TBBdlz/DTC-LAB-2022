# 4.1 การก่อกําเนิดตัวแปรสุ่ม

```julia
# define pdf of C
# f(C)
function f(a)
    if a>=0 && a<=2
        b=a/2
    else
        b=0
    end
    return b
end

# Plot pdf of R.V. C
using Plots
using LaTeXStrings
plotly()
x=range(-0.5,stop=2.5,length=1000)
c=f.(x)
plot(x,c,xlim=(-0.5,3),ylim=(0,1.2),label="")
plot!(title="pdf of random variable C",xlabel="C",ylabel="f(C)")
plot!(titlefont=font(12))
```

```julia
using Plots
using LaTeXStrings
plotly()
samples=5600
A=rand(samples)
C=2*sqrt.(A);
h=histogram(C,normed=true,nbins=60)
plot(h,xlim=(-0.5,3),ylim=(0,1.1),label="")
plot!(title="normalized histogram of random variable C",xlabel="C",ylabel="Histogram of C")
```

## 4.2 การก่อกําเนิดตัวแปรสุ่มแบบเกาส์

```julia
# generate Gaussian Random variable
# plot pdf pf Guassian Rv. mean 0
# variance =1
function f(a)
    b=1/(sqrt(2*pi)*sigma)*exp(-a^2/(2*sigma^2))
    return b
end

using Plots
using LaTeXStrings
plotly()
sigma=1
x=range(-4,stop=4,length=5600)
c=f.(x)
plot(x,c,xlim=(-4,4),ylim=(0,0.5),label="")
plot!(title="pdf of Gaussian random variable, C",xlabel="C",ylabel="f(C)")
```

```julia
# simulation of Gaussian random variable
# mean=0 variance=1
function gngauss(m,sigma)
    A=rand()
    r=sqrt(2*sigma^2*log(1/(1-A)))
    theta=2*pi*rand()
    u=m+r*cos(theta)
    v=m+r*sin(theta)
    return (u,v)
end

using Plots
using LaTeXStrings
pyplot()
m=0
sigma=1
samples=5600
C=zeros(samples)
for i=1:samples
    C[i],v=gngauss(m,sigma)
end
h=histogram(C,normed=true,nbins=60)
plot(h,xlim=(-4,4),ylim=(0,0.5),label="")
plot!(title="normalized histogram of Gaussian random variable C", xlabel="C",ylabel="Histogram of C")
```
